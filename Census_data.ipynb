```python
# feature Engineering 
### In feature engineering we need do categrorical column into numerical columns based on some assumption and some feature scaling  encoder
```

```python
data_copy.head()
```

```python
X=data_copy.drop(labels=["population"],axis=1)
Y=data_copy["population"]
```

```python
X.head()
```

```python
Y.head()
```

```python
cat_col=X.select_dtypes(include="object").columns
cat_col
```

```python
num_col=X.select_dtypes(exclude="O").columns
num_col
```

```python
workclass_cat=(' Never-worked',' Without-pay',' Self-emp-inc',' Local-gov',' Federal-gov',' State-gov',' Private', ' Self-emp-not-inc')

race_cat=(" Other"," Asian-Pac-Islander"," Amer-Indian-Eskimo"," Black"," White")

occupation_cat=(" Other-service"," Armed-Forces"," Handlers-cleaners"," Machine-op-inspct"," Craft-repair"," Farming-fishing",
                " Priv-house-serv"," Protective-serv"," Sales"," Transport-moving"," Adm-clerical"," Tech-support",
                " Prof-specialty"," Exec-managerial")
 
relation_cat=(" Not-in-family"," Other-relative"," Unmarried"," Own-child"," Wife"," Husband")

marital_cat=(" Widowed"," Separated"," Divorced"," Never-married",' Married-spouse-absent'," Married-AF-spouse"," Married-civ-spouse")

education_cat=(" Preschool", ' Prof-school'," 1st-4th"," 5th-6th"," 7th-8th"," 9th",' 10th'," 11th",' 12th'," Assoc-voc"," Assoc-acdm",
               " Some-college"," Bachelors"," HS-grad"," Doctorate"," Masters")

sex_cat=(" Female"," Male")

native_cat=(' United-States', ' Cuba', ' Jamaica', ' India', ' Mexico',' South', ' Puerto-Rico', ' Honduras',
           ' England', ' Canada',' Germany', ' Iran', ' Philippines', ' Italy', ' Poland',' Columbia', ' Cambodia', 
           ' Thailand', ' Ecuador', ' Laos',' Taiwan', ' Haiti', ' Portugal', ' Dominican-Republic',' El-Salvador', ' France',
           ' Guatemala', ' China', ' Japan',' Yugoslavia', ' Peru', ' Outlying-US(Guam-USVI-etc)', ' Scotland',
           ' Trinadad&Tobago', ' Greece', ' Nicaragua', ' Vietnam', ' Hong',' Ireland', ' Hungary', ' Holand-Netherlands')
```

```python
from sklearn.impute import SimpleImputer ## Handling Missing Values
from sklearn.preprocessing import StandardScaler # Handling Feature Scaling
from sklearn.preprocessing import OrdinalEncoder # Ordinal Encoding

## pipelines are use to -combining multiple steps

from sklearn.pipeline import Pipeline #use to combining multiple steps
from sklearn.compose import ColumnTransformer #for combining num and cat
```

```python
#for numerical:
num_pipeline=Pipeline(
    steps=[
    ("imputer",SimpleImputer(strategy="median")),
    ("scaler",StandardScaler())
])


# for categorical
cat_pipeline=Pipeline(
    steps=[
    ("imputer",SimpleImputer(strategy="most_frequent")),
    ("encoding",OrdinalEncoder(categories=[workclass_cat,education_cat,marital_cat,occupation_cat,relation_cat,race_cat,sex_cat,native_cat])),
    ("scaler",StandardScaler())
])


#pipeline

preprocessor=ColumnTransformer([
    ("num_pipeline",num_pipeline,num_col),
    ("cat_pipeline",cat_pipeline,cat_col)
])
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test=train_test_split(X,Y,test_size=0.30,random_state=42)
```

```python
X_train=pd.DataFrame(preprocessor.fit_transform(X_train))
X_test=pd.DataFrame(preprocessor.transform(X_test))
```

```python
X_train.head()
```

```python
X_test.head()
```

```python
# Now ,we need to take model,accuracy and all
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score,confusion_matrix
```

```python
model=LogisticRegression()
```

```python
model.fit(X_train,y_train)
```

```python
model.score(X_train,y_train)
```

```python
y_pred=model.predict(X_test)
```

```python
accuracy_score(y_test,y_pred)
```

```python
confusion_matrix(y_test,y_pred)
```
